```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
```

```python
#Landscape of Perspective-taking as described in Duran & Dale (2013)
k = .5 #Control parameter when participant believes that the partner is simulated
x= np.arange (-2, 2, 0.01)
V= k*x - (x**2)/2 + (x**4)/4

#Now plotting the potential function
plt.plot(x,V)
plt.xlabel("x")
plt.ylabel("Potential V")
plt.title("Believes Simulated")

#Duran and Dale (2013) propose that the initial condition for x in "believes simulated" condition is x= .2
#I'm therefore adding a point to that position for x
xpoint=.2
Vpoint= (.5*.2 - (.2**2)/2 + (.2**4)/4) + .06 #not an elegant way to do this, adding .06 to make the dot "sit" on the well. 
plt.scatter(xpoint,Vpoint, color ='red')


#the left valley represents the other-centric perspective as an attractor, 
#whereas the right one the egocentric perspective as an attractor

```

```python
#Landscape of Perspective-taking as described in Duran & Dale (2013)
k = .3 #Control parameter when participant believes that the partner is real
x= np.arange (-2, 2, 0.01)
V= k*x - (x**2)/2 + (x**4)/4

#Now plotting the potential function
plt.plot(x,V);

#Duran and Dale (2013) propose that the initial condition for x in "believes real" condition is x= .4
xpoint=.4
Vpoint= (.3*.4 - (.4**2)/2 + (.4**4)/4) + .06 #not an elegant way to do this, adding .06 to make the dot "sit" on the well. 
plt.scatter(xpoint,Vpoint, color ='red')

plt.xlabel("x")
plt.ylabel("Potential V")
plt.title("Believes Real")

#the left valley represents the other-centric perspective as an attractor, 
#whereas the right one the egocentric perspective as an attractor
```

```python
# Get a feel for parameter k
def f(k):
    V= k*x - (x**2)/2 + (x**4)/4
    plt.plot(x,V);
    plt.xlabel("x")
    plt.ylabel("Potential V")
    plt.title("Perspective Landscape")
interact(f, k= [-1,1,.1]);

#the left valley represents the other-centric perspective as an attractor, 
#whereas the right one the egocentric perspective as an attractor
```
